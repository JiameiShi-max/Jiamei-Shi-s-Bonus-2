```python
import gradio as gr

def fake_caption(image, creativity):
    # This is a mock function to demonstrate UI components
    if creativity < 5:
        return "A simple description of the image."
    else:
        return "A highly creative and artistic description of the image."

with gr.Blocks() as demo:
    gr.Markdown("# ðŸŽ¨ AI Creative Studio")

    with gr.Row():
        with gr.Column():
            image_input = gr.Image(type="pil", label="Upload an Image")
            creativity = gr.Slider(
                minimum=1,
                maximum=10,
                value=5,
                label="Creativity Level"
            )
            btn = gr.Button("Generate Caption")

        with gr.Column():
            caption_output = gr.Textbox(label="Generated Caption")

    btn.click(
        fn=fake_caption,
        inputs=[image_input, creativity],
        outputs=caption_output
    )

demo.launch()
```

```python
from transformers import pipeline
import gradio as gr

sentiment_pipe = pipeline("sentiment-analysis")

def analyze_sentiment(text):
    result = sentiment_pipe(text)[0]
    return f"Label: {result['label']}, Score: {result['score']:.2f}"

demo = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.Textbox(lines=2, placeholder="Enter a sentence..."),
    outputs=gr.Textbox(label="Sentiment Analysis Result"),
    title="ðŸ’¬ Sentiment Explorer",
    description="Analyze the sentiment of a sentence using a Hugging Face transformer model."
)

demo.launch()
```
